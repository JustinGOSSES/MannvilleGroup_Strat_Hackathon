```python
## Number 1 from brainstorm above
## Find average values of [a given curve from a given well] in [different] length windows [above, around, below] a depth
def findAvgWithinWindow(df,curve,depth,window,direction,UWI):
    ## establishing window top and bottom depth based on direction & window arguments
    if direction == 'above':
        top_depth = depth-window
        bottom_depth = depth
    elif direction == 'below':
        bottom_depth = depth+window
        top_depth = depth
    elif direction == "around":
        top_depth = depth - window/2
        bottom_depth = depth + window/2
    ##find the top_inx, bottom_inx from top and bottom depth and correct UWI
    top_inx = df.loc[(df['DEPT'] == top_depth) & (df['UWI'] == UWI)]
    bottom_inx = df.loc[(df['DEPT'] == bottom_depth) & (df['UWI'] == UWI)]
    ## check that both the top and bottom depth exist and we haven't gone above or below well coverage
    ## if we have..... we return None and don't do anything else. 
    ## this means None is the feature
    if top_inx.empty:
        print("couldn't print try 2, likely out of range")
        return None
    else:
#         print("try 2",df.loc[(df['DEPT'] == top_depth) & (df['UWI'] == UWI)])
#         print("try 2 2 , df_loc.index.values = ",df_loc.index.values[0])
        pass
    if bottom_inx.empty:
        print("couldn't print try 2b, likely out of range")
        return None
    else:
#         print("try 2",df_locB)
#         print("try 2 B , df_locB.index.values = ",df_locB.index.values[0])
        pass
    # finding the average value and returning it
    window_avg = np.average(df[[curve]][top_inx.index.values[0]:bottom_inx.index.values[0]+1])
    return window_avg
```

```python
#### Function that creates different average values for a range of arguments using the findAvgWithinWindow function
#### Returns an array of column names and values for diferent features
#### column names are constructed from curve, window length, and direction
####     for example: GR_2_Above
####     or: ILD_5_below
#### These in an array will be ["GR_2_Above":111.4,"GR_3_Above":161.4,"ILD_5_below":234]
def runFeatureFuncWithArgRange(df,rangeOfCurves,RangeOfWindows,RangeOfDirection,depth,UWI):
    # make a placeholder array for results
    results_placeholder = []
    ## Series of nested loops that output an array of new features for the average window function
    i=0
    for eachC in rangeOfCurves:
        for eachW in RangeOfWindows:
            for eachD in RangeOfDirection:
                window=str(eachW)
                col_name = 'avg'+'_'+eachC+'_'+'win'+window+'_'+eachD           
                results_placeholder.append({'colName':col_name,'value':findAvgWithinWindow(w_test_df,eachC,depth,eachW,eachD,UWI)})
                i=i+1
    return results_placeholder
```

```python
# Find average values of [a given curve from a given well] in [different] length windows [above, around, below] a depth
def findMaxMinPtsWithinWindow(df,curve,window,direction,depth,MinOrMax,NumbPts,UWI):
    # establishing window top and bottom depth based on direction & window arguments
    if direction == 'above':
        top_depth = depth-window
        bottom_depth = depth
    elif direction == 'below':
        bottom_depth = depth+window
        top_depth = depth
    elif direction == "around":
        top_depth = depth - window/2
        bottom_depth = depth + window/2
    #find the top_inx, bottom_inx from depths
#     top_inx=curve.index[curve['DEPT']==top_depth]
#     bottom_inx=curve.index[curve['DEPT']==bottom_depth]
    # get array of window
    #window_avg = np.average(curve[top_inx:bottom_inx+1])
    #window_avg = np.average(curve[top_depth:bottom_depth+1])
    
##### / added new    
    ##find the top_inx, bottom_inx from top and bottom depth and correct UWI
    top_inx = df.loc[(df['DEPT'] == top_depth) & (df['UWI'] == UWI)]
    bottom_inx = df.loc[(df['DEPT'] == bottom_depth) & (df['UWI'] == UWI)]
    ## check that both the top and bottom depth exist and we haven't gone above or below well coverage
    ## if we have..... we return None and don't do anything else. 
    ## this means None is the feature
    top_inx_values = top_inx.index.values[0]
    bottom_inx_values = bottom_inx.index.values[0]
    if top_inx.empty:
        print("couldn't print try 2, likely out of range")
        return None
    else:
#         print("try 2",df.loc[(df['DEPT'] == top_depth) & (df['UWI'] == UWI)])
#         print("try 2 2 , df_loc.index.values = ",df_loc.index.values[0])
        pass
    if bottom_inx.empty:
        print("couldn't print try 2b, likely out of range")
        return None
    else:
#         print("try 2",df_locB)
#         print("try 2 B , df_locB.index.values = ",df_locB.index.values[0])
        pass
    
##### added new / 

    if MinOrMax == 'max':
        orderedMax = np.maximum.accumulate(df[[curve]][top_inx_values:bottom_inx_values+1])
        ##print("orderedMax",orderedMax)
        if NumbPts < len(orderedMax):
            orderedMaxSmall = orderedMax[-NumbPts:]
            return np.average(orderedMaxSmall)
        else: 
            return None
    elif MinOrMax == 'min':
        orderedMin = np.minimum.accumulate(df[[curve]][top_inx_values:bottom_inx_values+1])
        if NumbPts < len(orderedMin):
            orderedMinSmall = orderedMin[-NumbPts:]
            return np.average(orderedMinSmall)
        else:
            return None
```

## FUNCTION FOR : Find difference between two windows on either side of a given depth. For each window, average for [different number of] [max, min] values in [different] length windows above & below a given depth for [each curve]

```python
#### Find difference between two windows on either side of a given depth. 
#### For each window, average for [different number of] [max, min] values in [different] length windows above & below a given depth for [each curve]
def findDiffMaxMinPtsBtweenWindows(df,curve,window,depth,MinOrMax,NumbPts,UWI):
    # establishing window top and bottom depth based on direction & window arguments
    top_depth = depth - window/2
    bottom_depth = depth + window/2
    #print("window =",window," NumbPts =",NumbPts)
    
    ##### / added new    
    ##find the top_inx, bottom_inx from top and bottom depth and correct UWI
    top_inx = df.loc[(df['DEPT'] == top_depth) & (df['UWI'] == UWI)]
    bottom_inx = df.loc[(df['DEPT'] == bottom_depth) & (df['UWI'] == UWI)]
    depth_inx = df.loc[(df['DEPT'] == depth) & (df['UWI'] == UWI)]
    ## check that both the top and bottom depth exist and we haven't gone above or below well coverage
    ## if we have..... we return None and don't do anything else. 
    ## this means None is the feature
    top_inx_values = top_inx.index.values[0]
    bottom_inx_values = bottom_inx.index.values[0]
    depth_inx_values = depth_inx.index.values[0]
    if top_inx.empty:
        print("couldn't print try 2, likely out of range")
        return None
    else:
    #         print("try 2",df.loc[(df['DEPT'] == top_depth) & (df['UWI'] == UWI)])
    #         print("try 2 2 , df_loc.index.values = ",df_loc.index.values[0])
        pass
    if bottom_inx.empty:
        print("couldn't print try 2b, likely out of range")
        return None
    else:
    #         print("try 2",df_locB)
    #         print("try 2 B , df_locB.index.values = ",df_locB.index.values[0])
        pass
    
    ##### added new / 
    NumbPts = int(NumbPts)
    
    #print("window,NumbPts",window,NumbPts)
    if MinOrMax == 'max':
        orderedMaxTop = np.maximum.accumulate(df[[curve]][top_inx_values:depth_inx_values+1])
        if NumbPts < len(orderedMaxTop):
            orderedMaxTopSmall = orderedMaxTop[-NumbPts:]
            averageTopMax = np.average(orderedMaxTopSmall)
        else:
            print("NumbPts and len(orderedMaxTop) ",NumbPts,len(orderedMaxTop))
            print("problem in max top")
        orderedMaxBottom = np.maximum.accumulate(df[[curve]][depth_inx_values:bottom_inx_values+1])
        if NumbPts < len(orderedMaxBottom):
            orderedMaxBottomSmall = orderedMaxBottom[-NumbPts:]
            averageBottomMax = np.average(orderedMaxBottomSmall)
        else:
            print("problem in max bottom")
            print("NumbPts and len(orderedMaxBottom) ",NumbPts,len(orderedMaxBottom))
        if NumbPts < len(orderedMaxBottom) and NumbPts < len(orderedMaxTop):
            dif_max_of_curve_at_depth = averageTopMax - averageBottomMax
            return dif_max_of_curve_at_depth
        else:
            print("problem in max both")  
    elif MinOrMax == 'min':
        orderedMinTop = np.minimum.accumulate(df[[curve]][top_inx_values:depth_inx_values+1])
        if NumbPts < len(orderedMinTop):
            orderedMinTopSmall = orderedMinTop[-NumbPts:]
            averageTopMin = np.average(orderedMinTopSmall)
        else:
            pass
        orderedMinBottom = np.minimum.accumulate(df[[curve]][depth_inx_values:bottom_inx_values+1])
        if NumbPts < len(orderedMinBottom):
            orderedMinBottomSmall = orderedMinBottom[-NumbPts:]
            averageBottomMin = np.average(orderedMinBottomSmall)
        else:
            pass
        if NumbPts < len(orderedMinBottom) and NumbPts < len(orderedMinTop):
            dif_min_of_curve_at_depth = averageTopMin - averageBottomMin
            return dif_min_of_curve_at_depth
        else:
            pass
    else:
        print(MinOrMax,"this shouldn't happen, failure in findDiffMaxMinPtsBtweenWindows, expected min or max as string"," NumbPts = ",NumbPts," and len(orderedMinBottom) =",len(orderedMaxTop))
```
